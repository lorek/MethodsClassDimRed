```python
import os
import pickle
import matplotlib.pyplot as plt
import sys

sys.path.append("..")
from models.linear_regression import myLinearRegression1D
```

```python
 
print(os.getcwd())
```

```python
# Specify the path to the pickle file
# Remember -- we a are in /notebooks
data_pkl_file = "../data/sample_data.pkl"
```

```python
# Check if the file exists
if os.path.exists(data_pkl_file):
    with open(data_pkl_file, "rb") as f:
        data = pickle.load(f)
```

```python


# Recover data from the loaded dictionary
x_data = data["x_data"]
y_data = data["y_data"]

model = myLinearRegression1D()
print("Fitting data...")
model.fit(x_data, y_data)
y_pred = model.predict(x_data)

# Plot the data
plt.figure(figsize=(7,4))
plt.scatter(x_data, y_data, alpha=0.3, label='Training data')
plt.plot(x_data, y_pred, color='red', label='Fitted regression line')
plt.xlabel("x")
plt.ylabel("y")
plt.title("1D Linear Regression Fit")
plt.legend()
plt.show()
```
